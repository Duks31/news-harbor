```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
def get_sentiment(text, model_name = "nlptown/bert-base-multilingual-uncased-sentiment"):
  tokenizer = AutoTokenizer.from_pretrained(model_name)
  model = AutoModelForSequenceClassification.from_pretrained(model_name)

  inputs = tokenizer(text, padding=True, truncation=True, return_tensors="pt", max_length=512)

  with torch.no_grad():
      outputs = model(**inputs)
  predictions = torch.argmax(outputs.logits, dim=1).item()

  sentiment_labels = {
      0: "Very Negative",
      1: "Negative",
      2: "Neutral",
      3: "Positive",
      4: "Very Positive"
  }
  return sentiment_labels[predictions]

```

```python
print(type(df['Description']))
```

```python
df['Sentiment'] = df[['Description']].apply(lambda x: get_sentiment(x[0]), axis=1)
```
